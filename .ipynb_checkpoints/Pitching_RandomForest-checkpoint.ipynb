```python
import pandas as pd
import numpy as np
```

```python
# IMPORT PITCHING DATA AND ASSIGN HOF AS 0 FOR 'NO' AND 1 FOR 'YES' 
pitching_df = pd.read_csv("Resources/pitcher_data.csv", index_col=0)
pitching_df.replace({'HOF': {'N':0, 'Y': 1}}, inplace=True)
pitching_df.head()
```

```python
# PRINT ORIGINAL DF SHAPE AND DISPLAY COUNTS OF HOF 
print(pitching_df.shape)
pitching_df['HOF'].value_counts()
```

```python
### THIS IS JUST FOR TESTING THE SPLIT BASED ON PANDAS FILTERS
test = pitching_df[(pitching_df['games'] > 100) ]
a = test['HOF'].value_counts()
split = (a[0] - a[1])/test.shape[0]
print(a)
print(split)
```

```python
# CELL USED TO DISPLAY MEAN, MIN, MAX OF A PARTICULAR COLUMN
# SOLELY FOR DATA REVIEW
d = 'games'

print("MAX " + d + ": " + str(pitching_df[d].max()))
print("MEAN " + d + ": " + str(pitching_df[d].mean()))
print("MIN " + d + ": " + str(pitching_df[d].min()))
```

```python
# DISPLAY COLUMN VALUES
pitching_df.columns
```

```python
# MORE THEN 100 GAMES PLAYED HAS BEEN DETERMINED TO DESIGNATE A PLAYER WHO HAS BEEN AN ESTABLISHED PITCHER
pitch = pitching_df[(pitching_df['games'] > 100)]
# HOF VALUE IS THE TARGET
target = pitch['HOF']
print(pitch.shape)
target_names = ['negative', 'positive']
data =pitch.drop('HOF', axis=1)

# THESE ARE THE FINAL COLUMNS DETERMINED TO BE OF PRIMARY VALUE
d = data[['ERA','wins','strikes','games','seasons','Opposing_BA']]

# SPLIT DATA INTO TEST AND TRAIN
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(d, target, random_state=42)
d.head()
```

```python
# PRINT SCORE 
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
# PRINT OUT THE CLASSIFICATION REPORT
from sklearn.metrics import classification_report
predictions = rf.predict(X_test)
print(classification_report(y_test, predictions,
                            target_names=["Positive","Negative"]))
```

```python
# RUN PREDICTIONS ON THE PRIMARY ANALYZED DATA
# ASSIGN THE PREDICTIONS TO A COLUMN VALUE IN THE DF
d['predictions'] = rf.predict(d)
```

```python
# CONCATENATE THE PREDICTED RESULTS ON THE INDEX
# THE PLAYERID CAN NOW BE VALIDATED WITH THE RESULTS
final_pitching = pd.concat([pitch, d], axis=1)
final_pitching.head()
```

```python
# CONVERT FINAL DF THAT CONTAINS ACTUAL AND PREDICTED HOF VALUES IN TO A CSV RESOURCE
final_pitching.to_csv('final_data/pitcher_dataFinal.csv')
```
